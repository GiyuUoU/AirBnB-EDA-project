#### Task 4. Data Cleaning

```python
data.isnull().sum()
```

```python
# Droping all missing values rows
data.dropna( inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with Duplicates
data.duplicated().sum()
data[data.duplicated()]
```

```python
# Deleting Duplicates
data.drop_duplicates( inplace=True)
data.duplicated().sum()
```

```python
# (Type casting) Changing the data type of 'id' column from int to object

data['id'] = data['id'].astype('object')
data.dtypes

data['host_id'] = data['host_id'].astype('object')
data.dtypes
```

#### Task 5. Data Analysis EDA
**Univariate Analysis**

```python
# Identifying Outliers in price column using boxplot

df = data[data['price'] < 1500]

sns.boxplot(data = df , x='price')
```

```python
# Price Distribution

plt.Figure(figsize=(8,5))
sns.histplot(data = df , x='price' , bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.dtypes
```

```python
# availability_365 Distribution

plt.Figure(figsize=(6,3))
sns.histplot(data = df , x='availability_365' )
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**


```python
# new column price per bed

df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

**Bi Variate Analysis**
- One variable dependency in another variable

```python
df.columns
```

```python
sns.barplot(data=df , x='neighbourhood_group' , y='price' , hue = 'room_type')
```

```python
# Number  of reviews and Price relationship

plt.figure(figsize=(8,5))
plt.title('Locatlity and Review Dependency')
sns.scatterplot(data= df , x='number_of_reviews' , y='price' , hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data = df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographic Distribution of AirBnB Listings

plt.figure(figsize=(10,7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographic Distribution of AirBnB Listings')
```

```python
df.dtypes
```

```python
# Heatmap - Correlation of one variable with others for numerical columns

corr = []

for col in df.columns:
    if df[col].dtype != 'object' and df[col].dtype != 'str':
        corr.append(col)

print(corr)
```

```python
corr_matrix =df[corr].corr()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(data=corr_matrix, annot=True) 
```
